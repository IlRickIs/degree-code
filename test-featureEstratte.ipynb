```python
df_complete = pd.concat([df, df_pos_labeled, df_neu_labeled],ignore_index=True, axis=0)
columns_to_drop = ['sentiment', 'subjectId', 'utteranceId', 'sex']
df_complete.drop(columns=columns_to_drop, inplace=True)
df_complete.shape
df_complete.head()
```

```python
nan_indices = df_complete[df_complete.isna().any(axis=1)].index
nan_indices
```

```python
def evaluate_model(model, x_test, y_test, averageM = 'macro'):
    from sklearn import metrics
    # Predict Test Data
    y_pred = model.predict(x_test)
    # Calculate accuracy, precision, recall, f1-score, and kappa score
    acc = metrics.accuracy_score(y_test, y_pred)
    prec = metrics.precision_score(y_test, y_pred, average = averageM)
    rec = metrics.recall_score(y_test, y_pred, average=averageM)
    f1 = metrics.f1_score(y_test, y_pred, average=averageM)
    kappa = metrics.cohen_kappa_score(y_test, y_pred, weights='quadratic')
    cm = metrics.confusion_matrix(y_test, y_pred)
    return {'acc': acc, 'prec': prec, 'rec': rec, 'f1': f1, 'kappa': kappa, 'cm': cm}
```

```python

from sklearn.model_selection import train_test_split
from sklearn.neighbors import  KNeighborsClassifier
```

```python
from sklearn.model_selection import train_test_split
#split dataset into train and test data

X_train, X_test, y_train, y_test = train_test_split(df_complete.iloc[:, :-1], df_complete['emotion'], test_size=0.2, random_state=42)
```

```python
# create a knn classifier algorithm with 30 neighbors
knn = KNeighborsClassifier(n_neighbors=30, metric='euclidean')
# fit the model
knn.fit(X_train, y_train)
```

```python
dtc_eval = evaluate_model(knn, X_test, y_test, averageM='macro')
dtc_eval
```

low perfomance, so we try to select the most significative features

```python
from sklearn.feature_selection import VarianceThreshold
X = df_complete.iloc[:, :-1]
sel = VarianceThreshold(threshold=(.8 * (1 - .8)))
X_filtered = sel.fit_transform(X)
X.shape, X_filtered.shape
```

we retry after this step of feature selection

```python
X_train, X_test, y_train, y_test = train_test_split(X_filtered, df_complete['emotion'], test_size=0.2, random_state=42)
neigh = KNeighborsClassifier(n_neighbors=30, metric='euclidean')
neigh.fit(X_train, y_train)
dtc_eval = evaluate_model(neigh, X_test, y_test, averageM='macro')
dtc_eval
```

no results? try again filtering deeper

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
```

```python
X_new = SelectKBest(f_classif, k=7).fit(X, df_complete['emotion'])
X_new.get_feature_names_out()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_new.transform(X), df_complete['emotion'], test_size=0.2, random_state=42)
neigh = KNeighborsClassifier(n_neighbors=30, metric='euclidean')
neigh.fit(X_train, y_train)
dtc_eval = evaluate_model(neigh, X_test, y_test, averageM='macro')
dtc_eval
```

now we try with parameter optimization

```python
from sklearn.model_selection import GridSearchCV
estimator_KNN = KNeighborsClassifier(algorithm='auto')
parameters_KNN = {
    'n_neighbors': (1,100, 1),
    'leaf_size': (10,100,1),
    'p': (1,2,3),
    'weights': ('uniform', 'distance'),
    'metric': ('minkowski', 'chebyshev', 'manhattan'),
}
                   
# with GridSearch
grid_search_KNN = GridSearchCV(
    estimator=estimator_KNN,
    param_grid=parameters_KNN,
    scoring = 'accuracy',
    n_jobs = -1,
    cv = 10
)

knn_optimized = grid_search_KNN.fit(X_train, y_train)
print(grid_search_KNN.best_params_)
dtc_eval = evaluate_model(knn_optimized, X_test, y_test, averageM='macro')
dtc_eval
```
